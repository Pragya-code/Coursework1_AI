(Outlier Detection)

3.1. Identify outlier countries in South Asia based on their Score and GDP per Capita.

3.2. Define outliers using the 1.5 x IQR rule.

```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")


south_df = df[df["Country name"].isin([
    "Afghanistan","Bangladesh","Bhutan","India",
    "Maldives","Nepal","Pakistan","Sri Lanka"
])].copy()

Q1 = south_df["score"].quantile(0.25)
Q3 = south_df["score"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

south_outliers = south_df[
    (south_df["score"] < lower) | (south_df["score"] > upper)
    ]
south_outliers
```

Task-3

3.3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers in a different color.

3.4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")


south_df = df[df["Country name"].isin([
    "Afghanistan","Bangladesh","Bhutan","India",
    "Maldives","Nepal","Pakistan","Sri Lanka"
])].copy()

gdp = [c for c in df.columns if "gdp" in c.lower()][0]

Q1 = south_df[gdp].quantile(0.25)
Q3 = south_df[gdp].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

south_outliers = south_df[(south_df[gdp] < lower_bound) | (south_df[gdp] > upper_bound)]

plt.figure(figsize=(8,5))
sns.scatterplot(data=south_df, x=gdp, y="score", label="Normal Countries")
sns.scatterplot(data=south_outliers, x=gdp, y="score", color="red", s=120, label="Outliers")

plt.title("Outliers in South Asia: GDP per Capita vs Happiness Score")
plt.xlabel("GDP per Capita")
plt.ylabel("Happiness Score")
plt.legend()
plt.show()

# The characteristics of these outliers and their potential impact on regional averages.
print("\n Outliers and Their potential impact on regional averages.")
print("\n - OUtliers are countries whose Score or GDP values are much higher or lower than others in the region.")
print("\n - Outliers can affect the average by pulling it up or down unfairly.")
print("\n - A country with very low GDP but average happiness may show that money is not the only factor.")
print("\n - A country with very low happiness can make the region look less happy overall.")
print("\n - Outliers show that South Asian countries are not all similar in happiness and economy.")
```

3.1. Compare key metice like GDP per Capita, Social Support, Healthy Life Expectancy between the regions using grouped bar charts.

3.2. Which metrics show the largest disparity betwen the two regions?

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")

south_asia = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
middle_east = [
    "Bahrain","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon","Oman",
    "Palestine","Qatar","Saudi Arabia","Syria","United Arab Emirates","Yemen"
]

south_df = df[df["Country name"].isin(south_asia)]
middle_df = df[df["Country name"].isin(middle_east)]

metrics = ["Log GDP per capita", "Social support", "Healthy life expectancy"]

south_means = south_df[metrics].mean()
middle_means = middle_df[metrics].mean()

x = np.arange(len(metrics))
width = 0.35

plt.figure(figsize=(10,6))
plt.bar(x - width/2, south_means, width, label="South Asia")
plt.bar(x + width/2, middle_means, width, label="Middle East")
plt.xticks(x, metrics)
plt.ylabel("Average Value")
plt.title("Metric Comparison: South Asia vs Middle East")
plt.legend()
plt.show()

print("\nLargest disparity metric:")
print((middle_means - south_means).abs().idxmax())
```

(Happiness Disparity:)

4.1. Compute the range (max - min) and coefficient of variation (CV) for Score in both regions.

4.2. Which region has greater variability in happiness ?


```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")

south_asia = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
middle_east = [
    "Bahrain","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon","Oman",
    "Palestine","Qatar","Saudi Arabia","Syria","United Arab Emirates","Yemen"
]

south_df = df[df["Country name"].isin(south_asia)]
middle_df = df[df["Country name"].isin(middle_east)]

# Range
south_range = south_df["score"].max() - south_df["score"].min()
middle_range = middle_df["score"].max() - middle_df["score"].min()

# Coefficient of Variation
south_cv = south_df["score"].std() / south_df["score"].mean()
middle_cv = middle_df["score"].std() / middle_df["score"].mean()

print("South Asia  Range:", south_range, "CV:", south_cv)
print("Middle East  Range:", middle_range, "CV:", middle_cv)

# To check which region has greater variability in happiness.
if middle_cv > south_cv:
    print("\nMiddle East has more variability in happiness.")
else:
    print("\nSouth Asia has more variability in happiness.")
```

Task-5

(Correlation Analysis:)

5.1. Analyze the correlation of Score with other metrics Freedom to Make Life Choices, and Generosity within each region.

5.2. Create scatter plots to visualize and interpret the relationships.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")

south_asia = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
middle_east = [
    "Bahrain","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon","Oman",
    "Palestine","Qatar","Saudi Arabia","Syria","United Arab Emirates","Yemen"
]

south_df = df[df["Country name"].isin(south_asia)]
middle_df = df[df["Country name"].isin(middle_east)]


# Correlations
print("South Asia Correlation:")
print(south_df[["score","Freedom to make life choices","Generosity"]].corr())

print("\nMiddle East Correlation:")
print(middle_df[["score","Freedom to make life choices","Generosity"]].corr())



# Scatter Plots
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.regplot(x=south_df["Freedom to make life choices"], y=south_df["score"])
plt.title("South Asia: Freedom vs Score")

plt.subplot(1,2,2)
sns.regplot(x=south_df["Generosity"], y=south_df["score"])
plt.title("South Asia: Generosity vs Score")

plt.show()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sns.regplot(x=middle_df["Freedom to make life choices"], y=middle_df["score"])
plt.title("Middle East: Freedom vs Score")

plt.subplot(1,2,2)
sns.regplot(x=middle_df["Generosity"], y=middle_df["score"])
plt.title("Middle East: Generosity vs Score")

plt.show()
```

Task-6

( Outlier Detection )

6.1. Identify outlier countries in both regions based on Score and GDP per Capita.

6.2. Plot these outliers and discuss their implications.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")

south_asia = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
middle_east = [
    "Bahrain","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon","Oman",
    "Palestine","Qatar","Saudi Arabia","Syria","United Arab Emirates","Yemen"
]

south_df = df[df["Country name"].isin(south_asia)].copy()
middle_df = df[df["Country name"].isin(middle_east)].copy()

def mark_outliers(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return df[(df[col] < lower) | (df[col] > upper)]

south_out = mark_outliers(south_df, "score")
middle_out = mark_outliers(middle_df, "score")

print("South Asia Score Outliers:")
print(south_out)

print("\nMiddle East Score Outliers:")
print(middle_out)

# Scatter plot
plt.figure(figsize=(10,6))
plt.scatter(south_df["Log GDP per capita"], south_df["score"], label="South Asia")
plt.scatter(middle_df["Log GDP per capita"], middle_df["score"], label="Middle East")
plt.scatter(south_out["Log GDP per capita"], south_out["score"], color="red", s=100, label="South Outliers")
plt.scatter(middle_out["Log GDP per capita"], middle_out["score"], color="orange", s=100, label="Middle Outliers")
plt.xlabel("GDP per Capita")
plt.ylabel("Score")
plt.legend()
plt.title("Outliers in South Asia & Middle East")
plt.show()
```

Task-7

( Visualization )

7.1. Create boxplots comparing the distribution of Score between South Asia and the Middle East.

7.2. Interpret the key differences in distribution shapes, medians, and outliers.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("/content/drive/MyDrive/Coursework1/WorldHappinessReport.csv")

south_asia = ["Afghanistan","Bangladesh","Bhutan","India","Maldives","Nepal","Pakistan","Sri Lanka"]
middle_east = [
    "Bahrain","Iran","Iraq","Israel","Jordan","Kuwait","Lebanon","Oman",
    "Palestine","Qatar","Saudi Arabia","Syria","United Arab Emirates","Yemen"
]

south_df = df[df["Country name"].isin(south_asia)].copy()
middle_df = df[df["Country name"].isin(middle_east)].copy()

# Combine for plotting
south_df["Region"] = "South Asia"
middle_df["Region"] = "Middle East"

combined = pd.concat([south_df, middle_df])

# Plot
plt.figure(figsize=(8,6))
sns.boxplot(data=combined, x="Region", y="score")
plt.title("Score Distribution: South Asia vs Middle East")
plt.show()
```
